# Corpus task pipeline

```python
from ekorpkit import eKonf
```

```python
corpus_cfg = eKonf.compose(config_group='corpus=corpus')
corpus_cfg.name = 'edgar'
corpus_cfg.automerge = True
corpus_cfg.data_dir = '/workspace/data/datasets/corpus/ekorpkit'
```

```python
cfg = eKonf.compose(config_group="pipeline")
cfg.data.corpus = corpus_cfg
cfg._pipeline_ = ["filter_query", "save_dataframe"]
ciks = [
    "320193",
    "789019",
    "1652044",
    "1018724",
    "200406",
    "104169",
    "80424",
    "9617",
    "70858",
    "21344",
]
cfg.filter_query.query = f"cik in {ciks}"
cfg.save_dataframe.output_dir = "../data/edgar"
cfg.save_dataframe.output_file = "edgar.parquet"
eKonf.instantiate(cfg)
```
